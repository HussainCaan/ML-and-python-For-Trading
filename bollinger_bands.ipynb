```python
import pandas as pd
import matplotlib.pyplot as plt

data = pd.read_excel("Ford data with MACD.xlsx")
data.set_index('Date', inplace=True)
data = data['9/1/2021':]

# moving average

ax = data['Close'].plot(title='Ford rolling mean', label='F')
rm_ford = data['Close'].rolling(window=7).mean()  # compute rolling mean
rstd_ford = data['Close'].rolling(window=7).std()  # compute rolling std
upper_band = rm_ford + (rstd_ford * 2)  # compute upper band
lower_band = rm_ford - (rstd_ford * 2)  # computer lower band

# plot raw ford values, mean and bollinger bands
rm_ford.plot(label='Rolling mean', ax=ax)
upper_band.plot(label='upper band', ax=ax)
lower_band.plot(label='lower band', ax=ax)

# Add axis label and lagend
ax.set_xlabel('Date')
ax.set_ylabel('Price')
ax.legend(loc='upper left')
plt.show()
```
