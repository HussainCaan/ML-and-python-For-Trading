```python
import pandas as pd
import matplotlib.pyplot as plt
Ford = pd.read_excel('Ford data with MACD.xlsx')
df = pd.DataFrame(Ford)
df.set_index('Date',inplace=True)
df = df['1/1/2021':]

close_prices = df['Close']

daily_returns = pd.Series(index=close_prices.index)
daily_returns.iloc[1:] = (close_prices.iloc[1:].values / close_prices.iloc[:-1].values) - 1
daily_returns.iloc[0] = 0

plt.hist(daily_returns,bins=20,edgecolor='black')

mean = daily_returns.mean()
std = daily_returns.std()

plt.axvline(mean,color = 'r')
plt.axvline(std,color = 'y')
plt.axvline(-std,color = 'y')
daily_returns.kurtosis()
```
